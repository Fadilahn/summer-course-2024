## First Case Study : Miles Per Gallon

```python
def MilesPerGallon():
    total_miles = 0
    total_gallons = 0
    tank_count = 0

    while True:
        gallons_used = float(input("Enter the gallons used (-1 to end): "))
        if gallons_used == -1:
            break

        miles_driven = float(input("Enter the miles driven: "))
        mpg = miles_driven / gallons_used

        print(f"The miles/gallon for this tank was {mpg:.6f}")

        total_miles += miles_driven
        total_gallons += gallons_used
        tank_count += 1

    if tank_count > 0:
        overall_average_mpg = total_miles / total_gallons
        print(f"\nThe overall average miles/gallon was {overall_average_mpg:.6f}")
    else:
        print("\nNo data entered.")
```

## Second Case Study : A Game of Chance - Crap

```python
import random

def roll_2dice():
    dice1 = random.randint(1, 6)
    dice2 = random.randint(1, 6)
    return dice1, dice2

def Craps():
    first = True
    while True:
        dice1, dice2 = roll_2dice()
        sum = dice1 + dice2
        print(f"Dice: {dice1 }, {dice2}")
        print(f"Sum: {sum}")

        if first:
            first = False
            if sum in (7, 11):
                print("You win!")
                break
            elif sum in (2, 3, 12):
                print("You lose!")
                break
            else:
                point = sum
                print(f"Point: {point}", end='\n\n')
        else:
            if sum == point:
                print("You make your point, You win!")
                break
            elif sum == 7:
                print("You rolled a 7, You lose!")
                break
```


RULES:
*   If the sum is 7 or 11 on the first roll, you win.
*   If the sum is 2, 3 or 12 on the first roll (called “craps”), you lose
*   If the sum is 4, 5, 6, 8, 9 or 10 on the first roll, that sum becomes your “point.”
*   To win, you must continue
rolling the dice until you “make your point” (i.e., roll that same point value)
*   You lose by rolling a 7 before
making your point.







```python
Craps()
```

## Third Case Study : COVID 19 Infection Statistic

```python
import pandas as pd

"""
WHO datasets:
Date_reported,Country_code,Country,WHO_region,New_cases,Cumulative_cases,New_deaths,Cumulative_deaths
"""

# Load the datasets from CSV file
file_path = '/content/WHO-COVID-19-global-data.csv'
df = pd.read_csv(file_path)
df.info()
```

```python
# Convert data type datetime
df['Date_reported'] = pd.to_datetime(df['Date_reported'])
df.head()
```

```python
# Filter rows for Indonesia
indonesia_data = df[df['Country'] == 'Indonesia']
```

```python
# Remove rows with NaN values in the ‘New_cases’ column
indonesia_data.dropna(subset=['New_cases'], inplace=True)
indonesia_data
```

```python
# @title Date_reported vs New_cases

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Date_reported']
  ys = series['New_cases']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = indonesia_data.sort_values('Date_reported', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Date_reported')
_ = plt.ylabel('New_cases')
```

```python
start_date = '2020-08-01'
end_date = '2021-01-31'

filtered_data = indonesia_data[(indonesia_data['Date_reported'] >= start_date)
& (indonesia_data['Date_reported'] <= end_date)]
filtered_data.head()
```

```python
def minimum(data):
    return min(data)

def maximum(data):
    return max(data)

def range(min_val, max_val):
    return max_val - min_val

def mean(data):
    return sum(data) / len(data)

def median(data):
    sorted_data = sorted(data)
    mid = len(sorted_data) // 2
    if len(sorted_data) % 2 == 0:
        return (sorted_data[mid - 1] + sorted_data[mid]) / 2
    else:
        return sorted_data[mid]

def variance(data, mean_val):
    squared_diff_sum = sum((x - mean_val) ** 2 for x in data)
    return squared_diff_sum / (len(data) - 1)

def std_dev(variance_val):
    return variance_val ** 0.5
```

```python
# infection_data = filtered_data['New_cases']
infection_data = [174, 335, 278, 214, 422, 513, 737, 672, 489, 412, 1301,
1105, 1123, 1376, 1502, 894, 665, 1704, 1656, 1342]
infection_data
```

```python
# Calculate statistics
min_infections = minimum(infection_data)
max_infections = maximum(infection_data)
range_infections = range(min_infections, max_infections)
mean_infections = mean(infection_data)
median_infections = median(infection_data)
variance_infections = variance(infection_data, mean_infections)
std_dev_infections = std_dev(variance_infections)

# Display the results
print(f"Minimum infections: {min_infections}")
print(f"Maximum infections: {max_infections}")
print(f"Range of infections: {range_infections}")
print(f"Mean infections: {mean_infections:.2f}")
print(f"Median infections: {median_infections:.2f}")
print(f"Variance of infections: {variance_infections:.2f}")
print(f"Standard deviation of infections: {std_dev_infections:.2f}")
```
